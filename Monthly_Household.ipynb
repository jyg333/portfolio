```python

"""5년간 업종별 소비 변화에 따른 주가변동 연관성을 조사.
주류 및 담배, 의류 및 신발, 교육비
각 소비가 차지하는 전체 비율 조사."""

data = pd.read_csv('expenditure01.csv')
# print(data.head())
pd.set_option('display.max_rows',data.shape[0]+1)
pd.get_option('display.max_columns',None)

# print(data.columns)
df = pd.DataFrame(data)

#각 요소의 위치를 파악하여 추출하기 위해 함수를 만들었다.
def getIndexes(df, value):
    list_of_Pos = []

    result = df.isin([value])

    #boolean series를 리턴
    series = result.any()

    # Get list of columns
    columns = list(series[series == True].index)

    # row index를 추출
    for col in columns:
        rows = list(result[col][result[col] == True].index)

        for row in rows:
            list_of_Pos.append((row, col))

    return list_of_Pos

# 요소 검색
x = str(input(print('Finding element: ' )))
list_of_Elems = [x]

position = {elem: getIndexes(df, elem) for elem in list_of_Elems}
print('데이터프레임에서 요소의 위치 : ')

# DataFrame애서 찾기
for value in position.items():
    print(value)


# print("index:",data.index[:])
#주류, 담배 데이터를 월별로 추출, A :alcohol/C: cigarette
#필요없는 데이터를 제외했다.
data_2015_AC = data.iloc[1:20,5]
data_2015_AC = data_2015_AC.drop([1,2,3,7,11,15,19])
data_2015_AC = data_2015_AC.reset_index(drop=True)

data_2016_AC = data.iloc[1:20,30]
data_2016_AC = data_2016_AC.drop([1,2,3,7,11,15,19])
data_2016_AC = data_2016_AC.reset_index(drop=True)

data_2017_AC = data.iloc[1:20,55]
data_2017_AC = data_2017_AC.drop([1,2,3,7,11,15,19])
data_2017_AC = data_2017_AC.reset_index(drop=True)

data_2018_AC = data.iloc[1:20,80]
data_2018_AC = data_2018_AC.drop([1,2,3,7,11,15,19])
data_2018_AC = data_2018_AC.reset_index(drop=True)

data_2019_AC = data.iloc[1:20,105]
data_2019_AC = data_2019_AC.drop([1,2,3,7,11,15,19])
data_2019_AC = data_2019_AC.reset_index(drop=True)
# print(data_2015_AC)

#C: clothes/ S : shoes
data_2015_CS = data.iloc[1:20,6]
data_2015_CS = data_2015_CS.drop([1,2,3,7,11,15,19])
data_2015_CS = data_2015_CS.reset_index(drop=True)

data_2016_CS = data.iloc[1:20,31]
data_2016_CS = data_2016_CS.drop([1,2,3,7,11,15,19])
data_2016_CS = data_2016_CS.reset_index(drop=True)

data_2017_CS = data.iloc[1:20,56]
data_2017_CS = data_2017_CS.drop([1,2,3,7,11,15,19])
data_2017_CS = data_2017_CS.reset_index(drop=True)

data_2018_CS = data.iloc[1:20,81]
data_2018_CS = data_2018_CS.drop([1,2,3,7,11,15,19])
data_2018_CS = data_2018_CS.reset_index(drop=True)

data_2019_CS = data.iloc[1:20,106]
data_2019_CS = data_2019_CS.drop([1,2,3,7,11,15,19])
data_2019_CS = data_2019_CS.reset_index(drop=True)
# print(data_2019_CS)
#E : Education
data_2015_E = data.iloc[1:20,13]
data_2015_E = data_2015_E.drop([1,2,3,7,11,15,19])
data_2015_E = data_2015_E.reset_index(drop=True)
# # print(data_2015_E)
data_2016_E = data.iloc[1:20,38]
data_2016_E = data_2016_E.drop([1,2,3,7,11,15,19])
data_2016_E = data_2016_E.reset_index(drop=True)
# # print(data_2016_E)
data_2017_E = data.iloc[1:20,63]
data_2017_E = data_2017_E.drop([1,2,3,7,11,15,19])
data_2017_E = data_2017_E.reset_index(drop=True)
# # print(data_2017_E)
data_2018_E = data.iloc[1:20,88]
data_2018_E = data_2018_E.drop([1,2,3,7,11,15,19])
data_2018_E = data_2018_E.reset_index(drop=True)
# # print(data_2018_E)
data_2019_E = data.iloc[1:20,113]
data_2019_E = data_2019_E.drop([1,2,3,7,11,15,19])
data_2019_E = data_2019_E.reset_index(drop=True)
# print(data_2019_E)

AC = pd.concat([data_2015_AC,data_2016_AC,data_2017_AC,data_2018_AC,data_2019_AC],keys=['2015','2016','2017','2018','2019'])
CS = pd.concat([data_2015_CS,data_2016_CS,data_2017_CS,data_2018_CS,data_2019_CS],keys=['2015','2016','2017','2018','2019'])
E = pd.concat([data_2015_E,data_2016_E,data_2017_E,data_2018_E,data_2019_E],keys=['2015','2016','2017','2018','2019'])
print(AC)
print(CS)
print(E)


#dtype: object 라서 plot 구현이 되지 않음
AC = pd.to_numeric(AC,errors='coerce')
CS = pd.to_numeric(CS,errors='coerce')
E = pd.to_numeric(E,errors='coerce')
```
